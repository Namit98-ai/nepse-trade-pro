```python
# ── Cell 2: Upload & extract project ────────────────────────────────────────
import os

# If you have the zip, unzip it:
# !unzip -q nepse_trade_pro.zip

# OR create structure from scratch:
os.makedirs('nepse_trade_pro/src', exist_ok=True)
os.makedirs('nepse_trade_pro/data', exist_ok=True)
os.makedirs('nepse_trade_pro/logs', exist_ok=True)

print('📁 Folder structure ready')
print('⚠️  Make sure you upload all .py files to nepse_trade_pro/ and nepse_trade_pro/src/')
```

```python
# ── Cell 3: Start Streamlit in background ───────────────────────────────────
import subprocess
import time

proc = subprocess.Popen(
    ['streamlit', 'run', 'app.py',
     '--server.port', '8501',
     '--server.headless', 'true',
     '--server.enableCORS', 'false'],
    cwd='nepse_trade_pro',
    stdout=subprocess.PIPE,
    stderr=subprocess.PIPE
)
time.sleep(4)
print('✅ Streamlit started on port 8501')
```

```python
# ── Cell 4: Get public URL via localtunnel ──────────────────────────────────
# This gives you a URL you can open on your phone!
import subprocess
result = subprocess.run(['curl', '-s', 'https://ipv4.icanhazip.com'],
                        capture_output=True, text=True)
ip = result.stdout.strip()
print(f'\n🌐 Your Colab external IP: {ip}')
print('If the tunnel asks for a password, paste this IP.\n')

# Start tunnel (URL appears below)
!npx localtunnel --port 8501
```
